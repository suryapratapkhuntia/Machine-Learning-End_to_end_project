```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/cars.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('selling_price', axis=1)
y = df['selling_price']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder()
```

```python
categorical_cols = x_train.select_dtypes(include='object').columns.tolist()
numerical_cols = x_train.select_dtypes(include=np.number).columns.tolist()

categorical_cols_test = x_test.select_dtypes(include='object').columns.tolist()
numerical_cols_test = x_test.select_dtypes(include=np.number).columns.tolist()

print("Categorical columns (train):", categorical_cols)
print("Numerical columns (train):", numerical_cols)
print("Categorical columns (test):", categorical_cols_test)
print("Numerical columns (test):", numerical_cols_test)
```

```python
x_train_encoded = oe.fit_transform(x_train[categorical_cols]).toarray()
x_test_encoded = oe.transform(x_test[categorical_cols_test]).toarray()
encoded_feature_names = oe.get_feature_names_out(categorical_cols)
```

```python
oe = OneHotEncoder(handle_unknown='ignore')
x_train_encoded = oe.fit_transform(x_train[categorical_cols]).toarray()
x_test_encoded = oe.transform(x_test[categorical_cols_test]).toarray()
encoded_feature_names = oe.get_feature_names_out(categorical_cols)
```

```python
x_train_encoded_df = pd.DataFrame(x_train_encoded, columns=encoded_feature_names, index=x_train.index)
x_train_processed = pd.concat([x_train_encoded_df, x_train[numerical_cols]], axis=1)

x_test_encoded_df = pd.DataFrame(x_test_encoded, columns=encoded_feature_names, index=x_test.index)
x_test_processed = pd.concat([x_test_encoded_df, x_test[numerical_cols_test]], axis=1)

display(x_train_processed.head())
display(x_test_processed.head())
```
