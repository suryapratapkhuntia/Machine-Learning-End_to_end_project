```python

numerical_cols = car.select_dtypes(include=np.number)


correlation_matrix = numerical_cols.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='make', y='price', data=car)
plt.title('Box Plot of Price by Car Make')
plt.xlabel('Car Make')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.lmplot(x='engine-size', y='price', data=car, height=6, aspect=1.5)
plt.title('Relationship between Price and Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

```python
sns.lmplot(x='city-mpg', y='highway-mpg',hue='make', data=car, height=6,fit_reg=False)
plt.title('Relationship between Highway MPG and Curb Weight')
plt.xlabel('city-mpg')
plt.ylabel('Highway MPG')
plt.show()
```

```python
pd.pivot_table(car,index=['body-style','num-of-doors'],values='normalized-losses').plot(kind='bar',figsize=(10,5))
plt.title('Average Normalized Losses by Body Style and Number of Doors')
plt.xlabel('Body Style')
plt.ylabel('Average Normalized Losses')
plt.show()
```
