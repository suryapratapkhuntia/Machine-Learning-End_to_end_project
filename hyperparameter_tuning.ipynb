````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

````

````python
df=pd.read_csv("/content/archive (19).zip")
df.sample(5)
````

````python
df.info()
````

````python
from sklearn.model_selection import train_test_split
````

````python
x_train,x_test,y_train,y_test=train_test_split(df.drop('target',axis=1),df['target'],test_size=0.2)
x_train.shape
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
rf=RandomForestClassifier()
gb=GradientBoostingClassifier()
lr=LogisticRegression()
````

````python
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
````

````python
rf=RandomForestClassifier(max_samples=0.75,random_state=42,oob_score=True)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
````

````python
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
y_pred=gb.predict(x_test)
accuracy_score(y_test,y_pred)
````

````python
from sklearn.svm import SVC
svm=SVC()
svm.fit(x_train,y_train)
y_pred=svm.predict(x_test)
accuracy_score(y_test,y_pred)
````

````python
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_score(y_test,y_pred)
````

````python
from sklearn.model_selection import GridSearchCV,cross_val_score
````

````python
np.mean(cross_val_score(RandomForestClassifier(max_samples=0.75),x_train,y_train,cv=10,scoring='accuracy'))
````

````python
n_estimators=[20,60,100,120]
max_samples=[0.25,0.5,0.75,1.0]
max_features=[0.25,0.5,0.75,1.0]
max_depth=[1,2,5,10]
````

````python
param_grid={'n_estimators':n_estimators,'max_samples':max_samples,'max_features':max_features,'max_depth':max_depth}
````

````python
rf_grid=GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,verbose=2,n_jobs=-1)
````

````python
rf_grid.fit(x_train,y_train)
````

````python
rf_grid.best_params_
````

````python
rf_grid.best_score_
````

** Random search cv**

````python
n_estimators=[20,60,100,120]
max_samples=[0.25,0.5,0.75,1.0]
max_features=[0.25,0.5,0.75,1.0]
max_depth=[1,2,5,10]
boostrap=[True,False]
min_samples_split=[2,5,10]
min_samples_leaf=[1,2,5,10]
````

````python
param_grid={'n_estimators':n_estimators,'max_samples':max_samples,'max_features':max_features,'max_depth':max_depth,'bootstrap':boostrap,'min_samples_split':min_samples_split,'min_samples_leaf':min_samples_leaf}
````

````python
from sklearn.model_selection import RandomizedSearchCV
````

````python
rf_grid=RandomizedSearchCV(estimator=rf,param_distributions=param_grid,cv=5,verbose=2,n_jobs=-1)
````

````python
rf_grid.fit(x_train,y_train)
````

````python
rf_grid.best_score_
````

````python
rf_grid.best_params_
````
