```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

x=np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
db=DBSCAN(eps=3,min_samples=2).fit(x)
db.labels_
```

Here 0 -> shows one cluster
     1 -> another cluster
     -1 -> noise point

```python
db=DBSCAN(eps=3,min_samples=3).fit(x)
db.labels_
```

```python
db=DBSCAN(eps=10,min_samples=2).fit(x)
db.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

x,_=make_moons(n_samples=200,noise=0.05)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
dbscan=DBSCAN(eps=0.3,min_samples=5)
clusters=dbscan.fit_predict(x)
dbscan.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=dbscan.labels_,cmap='rainbow')
plt.show()
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=5)
clusters=dbscan.fit_predict(x)

plt.scatter(x[:,0],x[:,1],c=dbscan.labels_,cmap='rainbow')
plt.show()
```

```python
from sklearn.datasets import make_circles

x,_=make_circles(n_samples=200,noise=0.05)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=5)
clusters=dbscan.fit_predict(x)

plt.scatter(x[:,0],x[:,1],c=dbscan.labels_,cmap='viridis',marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

 Advantages of DBSCAN
- No need to specify number of clusters: Unlike K-Means, DBSCAN does not require you to predefine the number of clusters.
- Detects clusters of arbitrary shape: It can find clusters that are not necessarily spherical or convex, making it ideal for real-world data with irregular patterns.
- Handles noise and outliers well: DBSCAN naturally identifies and excludes outliers as noise, which improves clustering quality.
- Works well with spatial data: Particularly effective for geospatial applications, image analysis, and anomaly detection.


 Disadvantages of DBSCAN
- Struggles with varying density: It may fail to detect clusters if they have significantly different densities, as the same eps and minPts parameters are applied globally.
- Parameter sensitivity: Choosing appropriate values for eps (neighborhood radius) and minPts (minimum points to form a cluster) can be tricky and dataset-dependent.
- Not ideal for high-dimensional data: In high-dimensional spaces, the concept of density becomes less meaningful due to the curse of dimensionality.
- Performance issues on large datasets: DBSCAN can be computationally expensive for very large datasets, especially without spatial indexing.
