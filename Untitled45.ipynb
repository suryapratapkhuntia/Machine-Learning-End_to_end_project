```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/archive (5).zip')

df.sample(10)
```

```python
df = df[['Wine','Alcohol','Malic.acid']]
df.sample(5)
```

```python
color_dict={1:'red',2:'blue',3:'green'}
sns.scatterplot(x=df['Alcohol'], y=df['Malic.acid'], hue=df['Wine'], palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Wine',axis=1),df['Wine'],test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
fig , (ax1,ax2)=plt.subplots(ncols=2,figsize=(10,4))
ax1.scatter(x_train['Alcohol'],x_train['Malic.acid'],color='blue',label='Train Data')
ax1.set_title('before scaling')
ax2.scatter(x_train_scaled['Alcohol'],x_train_scaled['Malic.acid'],color='blue',label='Train Data')
ax2.set_title('after scaling')
color_dict={1:'red',2:'blue',3:'green'}
sns.scatterplot(x=x_train['Alcohol'], y=x_train['Malic.acid'], hue=y_train, palette=color_dict, ax=ax1)
sns.scatterplot(x=x_train_scaled['Alcohol'], y=x_train_scaled['Malic.acid'], hue=y_train, palette=color_dict, ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))

# KDE plot for Alcohol before scaling
sns.kdeplot(data=x_train, x='Alcohol', hue=y_train, palette=color_dict, ax=ax1)
ax1.set_title('Alcohol Distribution Before Scaling')

# KDE plot for Alcohol after scaling
sns.kdeplot(data=x_train_scaled, x='Alcohol', hue=y_train, palette=color_dict, ax=ax2)
ax2.set_title('Alcohol Distribution After Scaling')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))

# KDE plot for Malic.acid before scaling
sns.kdeplot(data=x_train, x='Malic.acid', hue=y_train, palette=color_dict, ax=ax1)
ax1.set_title('Malic.acid Distribution Before Scaling')

# KDE plot for Malic.acid after scaling
sns.kdeplot(data=x_train_scaled, x='Malic.acid', hue=y_train, palette=color_dict, ax=ax2)
ax2.set_title('Malic.acid Distribution After Scaling')

plt.tight_layout()
plt.show()
```
