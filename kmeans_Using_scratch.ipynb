```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/student_clustering (1).csv')
df.sample(5)
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(df[['cgpa','iq']])
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**using scratch**

```python
class k_means_surya:
  def __init__(self, k=2, tol=0.001, max_iter=300):
    self.k = k
    self.tol = tol
    self.max_iter = max_iter


  def assign_cluster(self,x):
      cluster_group=[]
      distances=[]
      for i in range(len(x)):
        for centroid in self.centroids:
          distance=np.linalg.norm(x[i]-centroid)
          distances.append(distance)
        cluster_group.append(np.argmin(distances))
        distances=[]
      self.cluster_group=cluster_group
      return self.cluster_group

  def fit_predict (self,x):
    random_index=np.random.choice(len(x),self.k,replace=False)

    self.centroids=x.values[random_index,:]

    for i in range(self.max_iter):
      cluster_group=self.assign_cluster(x.values)
      old_centroid=self.centroids

      self.centroids=self.get_centroids(x, cluster_group)

      if (old_centroid == self.centroids).all():
        break

    return cluster_group

  def get_centroids(self,x,cluster_group):
    centroids=np.zeros((self.k,len(x.values[0])))

    for i in range(self.k):
      points=[x.values[j] for j in range(len(x)) if cluster_group[j]==i]
      centroids[i]=np.mean(points,axis=0)
    self.centroids=centroids
    return self.centroids
```

```python
surya=k_means_surya()
surya.fit_predict(df[['cgpa','iq']])
```
