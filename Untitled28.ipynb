```python
sns.scatterplot(x=df["cgpa"], y=df["iq"], hue=df['placement'])
plt.show()
```

```python
# if you saw a line can separate two parts then use logistic regression
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1:]
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
x_train
```

```python
x_test = scalar.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# seeing the line graph which is plotted
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train,y_train.values.ravel(),clf=clf,legend=2)
plt.show()
```
