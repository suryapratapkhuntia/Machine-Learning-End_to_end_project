```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
df=pd.DataFrame(x,columns=["x"])
df["y"]=y
df
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red")
```

```python
#lets apply gradiant Descent assuming slope is constant m = 78.35063668
#let assume starting value of b=1

y_pred=((78.35063668 * x) + 1).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label='ols')

plt.plot(x,y_pred,color="#00a65a",label='b=1')
plt.legend()
plt.show()
```

```python
m = 78.35
b=100

loss_slope=-2 * np.sum(y - (m * x.ravel() + b))
loss_slope
```

```python
learning_rate=0.1

step_size= learning_rate * loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((78.35063668 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label='ols')


plt.plot(x,y_pred1,color="green",label='b=100')
plt.legend()
plt.show()
```

```python
#itreation 2

m = 78.35
b=100

loss_slope=-2 * np.sum(y - (m * x.ravel() + b))


learning_rate=0.1

step_size= learning_rate * loss_slope


b=b-step_size
b
```

```python
y_pred2=((78.35063668 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label='ols')


plt.plot(x,y_pred1,color="green",label='b=100')
plt.plot(x,y_pred2,color="yellow",label='b=100')
plt.legend()
plt.show()
```

```python
#itreation 3

m = 78.35
b=100

loss_slope=-2 * np.sum(y - (m * x.ravel() + b))


learning_rate=0.1

step_size= learning_rate * loss_slope


b=b-step_size
b
```

```python
y_pred2=((78.35063668 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red",label='ols')


plt.plot(x,y_pred1,color="green",label='b=100')
plt.plot(x,y_pred2,color="yellow",label='b=100')
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
lr=0.01

epochs=100

for i in range(epochs):
    loss_slope=-2 * np.sum(y - (m * x.ravel() + b))
    b=b-lr * loss_slope
    print(b)

    y_pred=m * x + b
    plt.plot(x,y_pred)
    plt.scatter(x,y)
```
